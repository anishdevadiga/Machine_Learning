```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
data=load_iris()
X=data.data
y=data.target
target_names=data.target_names
```

```python
scaler=StandardScaler()
X_std=scaler.fit_transform(X)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_std)
```

```python
lda=LDA(n_components=2)
X_lda=lda.fit_transform(X_std,y)
```

```python
print('PCA Compements:\n',pca.components_)
print('LDA components:\n',lda.scalings_)
```

```python
plt.figure(figsize=(12,6))
#plotting PCA
plt.subplot(1,2,1)
for i,target_name in zip([0,1,2],target_names):
  plt.scatter(X_pca[y==i,0],X_pca[y==i,1],label=target_name)
plt.xlabel("PCA Component 1")
plt.ylabel("Pca component 2")
plt.title("PCA of IRIS dataset")
plt.legend()
```

```python
#plotting lda
plt.subplot(1,2,2)
for i,target_name in zip([0,1,2],target_names):
  plt.scatter(X_lda[y==i,0],X_lda[y==i,1],label=target_name)
plt.xlabel("LDA Component 1")
plt.ylabel("LDa component 2")
plt.title("LDA of iris dataset")
plt.legend()
plt.show()
```

```python
#variance ratio
print("Explained variance ratio by PCA ",pca.explained_variance_ratio_)
```
